```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
print(df.head())
```

```python
print(df.info())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.info())
```

```python
filtered_df = df.query("type == 'conventional' & region == 'Albany'")
filtered_df = filtered_df[filtered_df['Date'].dt.year == 2015]
filtered_df = filtered_df[['Date', 'Total Bags', 'Small Bags']]
print(filtered_df.head())
```

```python
melted_df = pd.melt(filtered_df, id_vars=['Date'], var_name='Bags', value_name='Count')
print(melted_df.head())
```

```python
sns.lineplot(data=melted_df, x='Date', y='Count', hue='Bags')
```

```python
df['Sales Volume'] = pd.qcut(df['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
print(df.head())
```

```python
df['Year'] = df['Date'].dt.year
sns.countplot(x='Year', hue='Sales Volume', data=df)
```

```python
bag_melted_df = pd.melt(df, id_vars=['Date'], value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], var_name='Bags', value_name='Count')
print(bag_melted_df.head())
```

```python
bag_melted_df['Year'] = bag_melted_df['Date'].dt.year
sns.relplot(data=bag_melted_df, kind='line', x='Year', y='Count', hue='Bags')
```

## Best Selling Bag Size
The `Small Bags` typically show the highest sales volume across the years, as indicated by the data analysis.
