# Exercise 3-1: Create some plots

```python
import pandas as pd
```

## Get the data

```python
mortality_data = pd.read_pickle('mortality_prepped.pkl')
mortality_data.head()
```

```python
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

## Visualize the data

```python
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
ax = mortality_wide['15-19 Years'].plot(kind='line', title='Mortality Rate for Age Group 15-19', legend=False, color='purple')
ax.set_ylabel('Mortality Rate')
```

```python
ax = mortality_wide.loc[2000].plot(kind='bar', title='Mortality Rates by Age Group in 2000')
ax.set_ylabel('Mortality Rate')
ax.set_xlabel('')
```

```python
ax = mortality_data.query("Year == 2000").pivot(index='AgeGroup', columns='Year', values='DeathRate').plot(kind='bar', legend=False, title='Mortality Rates by Age Group in 2000 Using Long Data')
ax.set_xlabel('')
ax.set_ylabel('Mortality Rate')
```

```python
ax = mortality_data['DeathRate'].plot(kind='hist', title='Histogram of Mortality Rates')

ax = mortality_data['DeathRate'].plot(kind='hist', bins=15, title='Histogram of Mortality Rates with 15 Bins')
```

```python
ax = mortality_data.query("Year == 1900")['DeathRate'].plot(kind='density', title='Density Plot of Mortality Rates in 1900')
ax.set_xlabel('Mortality Rate')
ax.grid(True)
```

```python
data = mortality_data.query('Year in [1900, 1925, 1950, 1975, 2000]').groupby('Year')['DeathRate'].sum()
ax = data.plot(kind='pie', title='Sum of Mortality Rates for Selected Years')
ax.set_ylabel('')
```

```python
years = [1900, 1925, 1950, 1975, 2000]

for year in years:
    ax = mortality_wide.loc[year].plot(kind='barh', title=f'Mortality Rates in {year}', figsize=(10, 8))
    ax.set_xlabel('Mortality Rate')
    ax.set_ylabel('Age Group')
    display(ax.figure)

```
